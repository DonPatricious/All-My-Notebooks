```python
#importing libraries
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

# Data Pre-process

```python
#load dataset
house_data = pd.read_csv('train.csv')
house_data.head()
```

```python
#checking the shape
house_data.shape
```

```python
#check for missing data
house_data.isnull().sum()
```

```python
#checking fro duplicates
house_data.duplicated().sum()
```

```python
house_data.describe()
```

# ML Model Creation

```python
#setting up target and features
#target
y = house_data.SalePrice

#feature
feature_col = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = house_data[feature_col]
```

```python
#creating test fitting
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
rf_house_test = RandomForestRegressor(random_state=1)
rf_house_test.fit(train_X, train_y)
rf_val_prediction = rf_house_test.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_prediction, val_y)
#testing predictions
print("The Mean Absolute Error is {:,.0f}".format(rf_val_mae))
```

```python
#introducing test data
#new model
rf_house = RandomForestRegressor(random_state=1)
rf_house.fit(X, y)
```

```python
#loading test data
house_test_data = pd.read_csv('test.csv')
```

```python
#create testing
test_X = house_test_data[feature_col]
test_prediction = rf_house.predict(test_X)
test_prediction
```

```python
#output result
output_data = pd.DataFrame({'ID':house_test_data.Id, 'SalePrice':test_prediction})
output_data.to_csv('output_data.csv', index=False)
```
