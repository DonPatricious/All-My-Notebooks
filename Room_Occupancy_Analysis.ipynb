## Exploratory Data Analysis

### Visualizing Temperature Fluctuations

```python
#plotting with temperature and occupancy combined
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(df['date'],
        df['Temperature'],
       color='#1B512D')
ax.set_ylabel('Temperature in degree Celsius', weight='bold')
ax1 = ax.twinx()
ax1.plot(df['date'],
         df['Occupancy'],
        color='#edae49')
ax1.set_ylabel('Occupancy', weight='bold')
ax1.set_yticks([0,1])
ax.set_xlabel('Date', weight='bold')
ax.set_title('Temperature Fluctuations and Occupancy',
            fontsize=14,
            weight='bold')
fig.legend(['Temperature', 'Occupancy'], fontsize='large')
plt.show()
```

### Visualizing Humidity Fluctuations

```python
#plotting with Humidity fluctuations and occupancy combined
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(df['date'],
        df['Humidity'],
       color='#1B512D')
ax.set_ylabel('Humidity in %', weight='bold')
ax1 = ax.twinx()
ax1.plot(df['date'],
         df['Occupancy'],
        color='#edae49')
ax1.set_ylabel('Occupancy', weight='bold')
ax1.set_yticks([0,1])
ax.set_xlabel('Date', weight='bold')
ax.set_title('Humidity Fluctuations and Occupancy',
            fontsize=14,
            weight='bold')
fig.legend(['Humidity', 'Occupancy'], fontsize='x-large')
plt.show()
```
