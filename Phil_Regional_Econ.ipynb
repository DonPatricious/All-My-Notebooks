```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_regions_of_the_Philippines_by_GDP'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0]
```

```python
table = soup.find_all('table')[0]
```

```python
table_headers = table.find_all('th')
```

```python
table_headers
```

```python
processed_table_headers = [titles.text.strip() for titles in table_headers]
processed_table_headers
```
