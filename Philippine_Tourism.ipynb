# Philippine Tourism

```python
import pandas as pd
```

```python
df = pd.read_excel('Regional_tourism_data_raw.xlsx')
```

```python
df
```

```python
df.info()
```

```python
df['Foreign Travelers'] = df['Foreign Travelers'].replace('-', '0')
df['Overseas Filipinos'] = df['Overseas Filipinos'].replace('-', '0')
df['Domestic Travelers'] = df['Domestic Travelers'].replace('-', '0')
df['Total'] = df['Total'].replace('-', '0')
```

```python
df
```

```python
#convert to numeric
df['Foreign Travelers'] = pd.to_numeric(df['Foreign Travelers'])
df['Overseas Filipinos'] = pd.to_numeric(df['Overseas Filipinos'])
df['Domestic Travelers'] = pd.to_numeric(df['Domestic Travelers'])
df['Total'] = pd.to_numeric(df['Total'])
```

```python
df.info()
```

```python
df
```

```python
#checking the provinces
df_provinces = pd.read_excel('Philippine Geography.xlsx', sheet_name='PROVINCE')
```

```python
df_provinces
```

```python
len(df)
```

```python
df['Province'].isin(df_provinces['Subdivision Name']).sum()
```

There are provinces that are not on the list

```python
#view provinces
df.loc[~df['Province'].isin(df_provinces['Subdivision Name'])]
```

```python
# changing the Mindoros names
df['Province'] = df['Province'].replace('Occidental Mindoro', 'Mindoro Occidental')
df['Province'] = df['Province'].replace('Oriental Mindoro', 'Mindoro Oriental')
```

```python
#view provinces
df.loc[~df['Province'].isin(df_provinces['Subdivision Name'])]
```

```python
#zamboangas
df['Province'] = df['Province'].replace('Zamboanga Del Norte', 'Zamboanga del Norte')
df['Province'] = df['Province'].replace('Zamboanga Del Sur', 'Zamboanga del Sur')
df['Province'] = df['Province'].replace('Zambonaga del Sur', 'Zamboanga del Sur')
```

```python
#davaos
df['Province'] = df['Province'].replace('Davao De Oro', 'Davao de Oro')
df['Province'] = df['Province'].replace('Davao Del Norte', 'Davao del Norte')
df['Province'] = df['Province'].replace('Davao Occcidental', 'Davao Occidental')
```

```python
#cotabatos
df['Province'] = df['Province'].replace('North Cotabato', 'Cotabato')
```

```python
#view provinces
df.loc[~df['Province'].isin(df_provinces['Subdivision Name'])]
```

```python
#export to excel
df.to_excel('Regional_tourism_data_processed.xlsx', index=False)
```
